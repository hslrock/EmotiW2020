```python
def load_pretrained_model(pre_train):
    frame_num=10
    audio_model=video_model.AudioRecognition(softmax=pre_train)
    image_model=video_model.Video_modeller(frame_num,pre_train=pre_train)
    return(image_model,audio_model)

img_model,audio_model=load_pretrained_model(False)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
torch.set_default_tensor_type(torch.cuda.FloatTensor)


model=video_model.video_transformer(img_model,audio_model)
model=model.to(device)
model.load_state_dict(torch.load("v7_model/"+'v7_37.pth'),strict=False)
```

```python
def show_frame(data):
    import matplotlib.pyplot as plt
    input=data
    plt.figure(figsize=(20, 20))
    grid_border_size = 2
    grid = utils.make_grid(input)

    plt.imshow((grid.numpy().transpose((1, 2, 0))*0.5)+0.5)
    plt.axis('off')
    plt.ioff()
    plt.show()
```

```python
def show_face(data):
    import matplotlib.pyplot as plt
    input=data[0]
    plt.figure(figsize=(20, 20))
    grid_border_size = 2
    grid = utils.make_grid(input)

    plt.imshow((grid.numpy().transpose((1, 2, 0))*0.5)+0.5)
    plt.axis('off')
    plt.ioff()
    plt.show()
```
